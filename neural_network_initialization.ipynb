    - In this file, contains the experiments of weight unitilization and its effect deep networks
    - here we are taking 10 layer neural network with each layer contains 512 neurons
    - experiments were performed on tanh and relu activation

```
#this file contains some of the visualization of the weight intialization
import numpy as np
import matplotlib.pyplot as plt


def forward(activation_name, input):
    """
    :param input: data from the direct input or previous layer
    :return:  forward pass of the Liner layer
    """


    if activation_name == "tanh":

        out = np.tanh(input)
        return out

    if activation_name == "relu":

        out = np.maximum(input, 0)
        return out

    if activation_name == "sigmoid":

        out = 1 / (1 + np.exp(input * -1))
        return out

    if activation_name == "linear":
        out = input
        return out

    if activation_name == "softmax":

        out = np.exp(input)
        out = out / (out.sum(axis=0) + 1e-10)
        return out

```

# Uniform Intialization
Here, by applying uniform distributurion for initalization, the varience is shifted to zero, which leads to dead network problem.

```
#uniform intitialization
features = 512
X = np.random.normal(size=(1000,features))
weights = [512]*10
activation = ['tanh']*10
w = np.random.randn(10,512,512)# intialization of weights

fig, a = plt.subplots(2, 5)

input = X
Distribution = [X]

fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=1.05)
for i in range(len(w)):
    a[int(i / 5)][i % 5].hist(input.ravel(),30,range = (-1,1))

    input = forward(activation[i], np.dot(input,w[i]))
    print(f"hidden layer number : {i+1} | mean : {input.mean()} | std : {input.std()}")
    Distribution.append(input)

plt.title(f"layer output distribution ({activation[0]})")

plt.show()

mean = [[x.mean()] for x in Distribution]
var =  [[x.std()] for x in Distribution]
plt.plot(mean)
plt.title(f"means shift ({activation[0]})")
plt.show()

plt.plot(var)
plt.title(f"varience shift ({activation[0]})")
plt.show()
```
